```python
# nahranie datasetu cez API 
import eurostat 

df = eurostat.get_data_df('gov_10a_exp')
df
```

```python
for column in df.columns:
    print(f"Unikátne hodnoty v stĺpci {column}:")
    print(df[column].unique())
    print("\n")
```

```python
#CHYBA!!!!! nemôžem ho odstrániť, musím si najprv len vyfiltrovať to kde je total
import pandas as pd
# odstránim stlpec cofog99, ktorý nám netreba
#df = df.drop(columns=['cofog99'])
#df

# Filtrovanie riadkov, kde v 'unit' nie je 'MIO_NAC'
df = df[df['cofog99'] == 'TOTAL']

# Zobrazenie výsledkov
df

#až teraz môžem odstrániť
df = df.drop(columns=['cofog99'])
df
```

```python
#chcem zistiť, čo ešte môžem vymazať - zobrazím si unikátne hodnoty v každom stlpci a zisťujem, že vo freq je len 1 - mažem stlpec
print(df.nunique())

# Zobrazenie konkrétnych unikátnych hodnôt pre každý stĺpec
for column in df.columns:
    print(f"Unikátne hodnoty v stĺpci {column}:")
    print(df[column].unique())
    print("\n")
```

```python
#odstránenie stlpca freq
df = df.drop(columns=['freq'])
df
```

```python
#premenovanie stlpca na country_code - to by malo byť všade
df = df.rename(columns={"geo\TIME_PERIOD": "country_code"})
df
```

```python
#Unpivotovanie dát pomocou melt
df_melted = pd.melt(df, 
                    id_vars=['unit', 'sector', 'na_item', 'country_code'],  #Stĺpce, ktoré chceme ponechať
                    var_name='year',   #Nový stĺpec, kde budu vsetky roky
                    value_name='value'  #Nový stĺpec pre hodnoty z rokov - teda tá výška investície buď ako percento alebo v milionoch eur
                   )

# Zobrazenie výsledkov
df_melted

```

```python
#ešte som zabudla na stlpec na_item, pozriem sa, čo v ňom je
unique_values = df['na_item'].unique()

#vytlačenie unikátnych hodnôt
print(unique_values)

#zistím, že chcem len TE - total government expenditure
```

```python
#filtrovanie riadkov, kde v stĺpci 'na_item' je hodnota 'TE'
#to isté spravím pre stlpec sector, kde zistím, že chcem len S13 (alebo sa to dá čeknúť aj na eurostat webovom datasete)
#tu si už vyfilturujem, že v stlpci sector chcem len S13 (general government) - je to sporné, lebo aj local governments investujú, ale whatever, nás zaujíma len general
filtered_df = df_melted[(df_melted['na_item'] == 'TE') & (df_melted['sector'] == 'S13')] 

#Zobrazenie výsledkov
filtered_df
```

```python
# Filtrovanie riadkov, kde v 'unit' nie je 'MIO_NAC'
prefinal_df = filtered_df[filtered_df['unit'] != 'MIO_NAC']

# Zobrazenie výsledkov
prefinal_df
```

```python
#ROZHODLA SOM SA, ŽE NEDÁM PREČ NULOVE HODNOTY
#na záver ešte dám preč NaN hodnoty zo stlpca value
#final_df= prefinal_df.dropna(subset='value')
#final_df

prefinal_df
```

```python
#Zobrazenie počtu unikátnych hodnôt v jednotlivých stĺpcoch, uistím sa, že mám všade, to, čo chcem
print(prefinal_df.nunique())

#Zobrazenie konkrétnych unikátnych hodnôt pre každý stĺpec
for column in prefinal_df.columns:
    print(f"Unikátne hodnoty v stĺpci {column}:")
    print(prefinal_df[column].unique())
    print("\n")
```

```python
#odstránenie stlpcov sector a na_item. V sector mám len to S13, čo je šport a niečo - je to na eurostate a v na_item mám len TE, čiže tie celkové vládne výdavky
prefinal_df = prefinal_df.drop(columns=['sector', 'na_item'])
prefinal_df
```

```python
prefinal_df['value'] = prefinal_df['value'].astype(float)
```

```python
prefinal_df.info()
```

```python
prefinal_df['value'] = prefinal_df['value'].astype(str).str.replace('.', ',', regex=False)
```

```python
prefinal_df['value'] = prefinal_df['value'].astype(str).str.replace('nan', '0')
```
