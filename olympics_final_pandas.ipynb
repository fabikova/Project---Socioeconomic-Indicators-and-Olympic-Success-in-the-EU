```python
#otvorenie datasetu - výsledky olympiády
import pandas as pd
df = pd.read_csv('olympic_results.csv')
df = pd.read_csv('olympic_results.csv')
df
```

```python
#výber stlpcov, ktoré chceme ponechať
selected_columns = [
    'discipline_title', 'slug_game', 'medal_type', 
    'rank_position', 'country_name', 'country_code', 'country_3_letter_code'
]
cleaned_df = df[selected_columns]
cleaned_df
```

```python
#chceme získať samostatny stlpec year a stlpec place
#Rozdelenie stĺpca slug_game na dva nové stĺpce place a year
cleaned_df[['place', 'year']] = cleaned_df['slug_game'].str.split('-', expand=True)

#Zmazanie pôvodného stĺpca slug_game
cleaned_df = cleaned_df.drop(columns=['slug_game'])

#Prehodenie stĺpcov tak, aby nové stĺpce boli na správnom mieste
cleaned_df = cleaned_df[['place', 'year'] + selected_columns[0:3] + selected_columns[4:]]

#zistila som, že niektoré stlpce majú hodnoty, ktoré nie sú oddelené pomlčkou, sucks... treba zistiť, čo s tým
```

```python
# Nájdeme riadky, ktoré nemajú presne jeden výskyt pomlčky
invalid_rows = cleaned_df[cleaned_df['slug_game'].str.count('-') != 1]
print(invalid_rows)
#zisťujeme, že niektoré mestrá majú viac pomlčiek
```

```python
#Extrahovanie roku (štvormiestneho čísla) zo stĺpca slug_game - regulérny výraz(regex), výsledok extrakcie bude uložený do nového stlpca year
cleaned_df['year'] = cleaned_df['slug_game'].str.extract(r'(\d{4})')


#Extrahovanie miesta (všetko pred rokom), loc je na prístup k riadkom a stĺpcom v df. dvojbodka (:) označuje, že chcem prístup ku všetkým riadkom
cleaned_df.loc[:, 'place'] = cleaned_df['slug_game'].str.replace(r'-\d{4}', '', regex=True) 

cleaned_df

#je tu nejaké upozornenie, ale neviem, čo s tým, lebo keby tam dám .loc aj k roku, tak to do year napíše "NaN" - ignorujeme to

```

```python
#overila som si, že skutočne mám pekne upravené roky v stlpci year
print(cleaned_df['year'].unique())
```

```python
#zistila som, že nemáme stlpec s tým, či ide o zimnú alebo letnú olympiádu, treba to spojiť s druhou tabuľkou olympic_hosts, kde je samostatný stlpec na season

# Načítanie novej CSV tabuľky
games_df = pd.read_csv('olympic_hosts.csv')

# Spojenie oboch tabuliek na základe game_slug - toto je stlpec, v ktorom majú rovnaké hodnoty
merged_df = pd.merge(cleaned_df, games_df, left_on='slug_game', right_on='game_slug', how='left')

# Zobrazenie výslednej tabuľky
merged_df
```

```python
#výber stlpcov, ktoré chceme ponechať, keďže tu máme zbytočne opakujúce sa stlpce
selected_columns_final = [
    'discipline_title', 'medal_type', 
    'rank_position', 'country_name', 'country_code', 
    'country_3_letter_code', 'game_slug', 
    'game_location', 'game_name', 'game_season', 'place', 'year'
]
merged_df_final = merged_df[selected_columns_final]
merged_df_final
```

```python
#export do csv
merged_df_final.to_csv("olympic_data_final.csv")
```
