```python
import pandas as pd
import eurostat
df = eurostat.get_data_df('hlth_ehis_pe9e')
df
```

```python
#zobrazenie unikátnych hodnôt

print(df.nunique())

# Zobrazenie konkrétnych unikátnych hodnôt pre každý stĺpec
for column in df.columns:
    print(f"Unikátne hodnoty v stĺpci {column}:")
    print(df[column].unique())
    print("\n")
```

```python
#mažem freq, lebo je tam len jedna a to A
df = df.drop(columns=['freq'])
df
```

```python
#z isced11 chcem vyfiltrovať len tie hodnoty, kde isced11 = TOTAL
df = df[df['isced11'] == 'TOTAL']
df
```

```python
#z age chcem vyfiltrovať len tie hodnoty, kde age = TOTAL
df = df[df['age'] == 'TOTAL']
df
```

```python
#premenujem stlpec na country_code
df = df.rename(columns={"geo\TIME_PERIOD": "country_code"})
df
```

```python
#mažem age, lebo už viem, že tam má len total, to isté isced11
df = df.drop(columns=['age', 'isced11'])
df
```

```python
#Unpivotovanie pomocou melt
df_melted = pd.melt(df, 
                    id_vars=['unit', 'physact', 'sex', 'country_code'],  #Stĺpce, ktoré chceme ponechať
                    var_name='year',   #Nový stĺpec, kde budu vsetky roky
                    value_name='value'  #Nový stĺpec pre hodnoty z rokov - teda tá výška investície buď ako percento alebo v milionoch eur
                   )

# Zobrazenie výsledkov
df_melted
```

```python
df_melted['value'] = df_melted['value'].astype(float)

df_melted['value'] = df_melted['value'].astype(str).str.replace('.', ',')

df_melted['value'] = df_melted['value'].astype(str).str.replace('nan', '0')
```

```python
#export tabuľky do csv
df_melted.to_csv('performing_physical_activity_final.csv', index= False)
```
