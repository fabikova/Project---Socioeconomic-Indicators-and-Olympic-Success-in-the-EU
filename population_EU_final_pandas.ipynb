```python
eu_and_related_countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 
    'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland', 
    'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 
    'Poland', 'Portugal', 'Romania', 'Slovak Republic', 'Slovenia', 'Spain', 
    'Sweden', 'United Kingdom', 'Turkey', 'Switzerland', 'Norway', 'Iceland', 
    'Liechtenstein', 'Bosnia and Herzegovina', 'Serbia', 'Montenegro', 
    'North Macedonia', 'Albania', 'Kosovo'
]


# Načítanie CSV súboru
file_path = "API_SP.POP.TOTL_DS2_en_csv_v2_31753.csv"
df = pd.read_csv(file_path, skiprows=4)

# Filtrovanie krajín podľa EÚ a príbuzných krajín
df_filtered = df[df['Country Name'].isin(eu_and_related_countries)]

# Zobrazenie výsledkov
df_filtered
```

```python
#odstránenie stlpcov, kde sú všade rovnaké hodnoty
df_filtered = df_filtered.drop(columns=['Indicator Name', 'Indicator Code'])
df_filtered
```

```python
#Unpivotovanie dát pomocou melt
df_melted = pd.melt(df_filtered, 
                    id_vars=['Country Name', 'Country Code'],  #Stĺpce, ktoré chceme ponechať
                    var_name='year',   #Nový stĺpec, kde budu vsetky roky
                    value_name='value'  #Nový stĺpec pre hodnoty z rokov - teda tá výška investície buď ako percento alebo v milionoch eur
                   )

# Zobrazenie výsledkov
df_melted
```

```python
#odstránenie divnej hodnoty 
df_melted = df_melted[df_melted['year'] != 'Unnamed: 68']
df_melted
```

```python

#dictionary na mapovanie trojpísmenových kódov na dvojpísmenové (ISO 3166-1 alpha-2)
country_code_map = {
    'ALB': 'AL', 'AUT': 'AT', 'BEL': 'BE', 'BGR': 'BG', 'BIH': 'BA', 'CHE': 'CH',
    'CYP': 'CY', 'CZE': 'CZ', 'DEU': 'DE', 'DNK': 'DK', 'ESP': 'ES', 'EST': 'EE',
    'FIN': 'FI', 'FRA': 'FR', 'GBR': 'GB', 'GRC': 'GR', 'HRV': 'HR', 'HUN': 'HU',
    'IRL': 'IE', 'ISL': 'IS', 'ITA': 'IT', 'LIE': 'LI', 'LTU': 'LT', 'LUX': 'LU',
    'LVA': 'LV', 'MKD': 'MK', 'MLT': 'MT', 'MNE': 'ME', 'NLD': 'NL', 'NOR': 'NO',
    'POL': 'PL', 'PRT': 'PT', 'ROU': 'RO', 'SRB': 'RS', 'SVK': 'SK', 'SVN': 'SI',
    'SWE': 'SE', 'XKX': 'XK'
}

# pridanie nového stĺpca country_code_2_letters na základe existujúceho slovníka
df_melted['country_code'] = df_melted['Country Code'].map(country_code_map)

# zobrazenie celej tabuľky aj s novým stĺpcom
df_melted

```

```python
# premenovanie stĺpcov na country_3_letter_code a country_name
df_melted.rename(columns={'Country Name': 'country_name', 'Country Code': 'country_3_letter_code'}, inplace=True)

# zobrazenie výslednej tabuľky
df_melted

```

```python
# získanie názvov stĺpcov
columns_names = df_melted.columns.tolist()

#presunutie posledného (piateho) stĺpca na tretie miesto (index 2)
columns_names.insert(2, columns_names.pop(-2))

#zmena poradia stĺpcov
df_melted = df_melted[columns_names]

# zobrazenie výslednej tabuľky
df_melted
```

```python
df_melted['value'] = df_melted['value'].astype(float)

df_melted['value'] = df_melted['value'].astype(str).str.replace('.', ',')

df_melted['value'] = df_melted['value'].astype(str).str.replace('nan', '0')
```

```python
#konečne export do csv
df_melted.to_csv("population_EU_final.csv", index=False)
```
