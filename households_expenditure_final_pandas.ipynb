```python
# načítanie dát cez API
import pandas as pd
import eurostat
df = eurostat.get_data_df('sprt_pcs_hbs')
df
```

```python
#kontrola hodnôt
print(df.nunique())

# Zobrazenie konkrétnych unikátnych hodnôt pre každý stĺpec
for column in df.columns:
    print(f"Unikátne hodnoty v stĺpci {column}:")
    print(df[column].unique())
    print("\n")

#usudzujem, že môžem zmazať stlpce freq, unit nechám, aby som vedela v akej to je jednotke
#coicop je že rôzne typy výdavkov ako na outdoor potreby, equipment, total atd., to tiež nechám
```

```python
#odstránenie stlpca freq
df = df.drop(columns=['freq'])
df
```

```python
#premenovanie stlpca na country_code
df = df.rename(columns={"geo\TIME_PERIOD": "country_code"})
df
```

```python
#unpivotovanie
df_melted = pd.melt(df, 
                    id_vars=['coicop', 'unit', 'country_code'],  #Stĺpce, ktoré chceme ponechať
                    var_name='year',   #Nový stĺpec, kde budu vsetky roky
                    value_name='value'  #Nový stĺpec pre hodnoty z rokov - teda tá výška výdavkov na konkrétne športové potreby
                   )

# Zobrazenie výsledkov
df_melted
```

```python
df_melted['value'] = df_melted['value'].astype(str).str.replace('.', ',')
```

```python
df_melted['value'] = df_melted['value'].astype(str).str.replace('nan', '0')
```

```python
for column in df_melted.columns:
    print(f"Unikátne hodnoty v stĺpci {column}:")
    print(df_melted[column].unique())
    print("\n")
```

```python
#rozhodla som sa nevymazať NaN hodnoty, keď tak to poriešime potom, ale nech máme akoby za každý rok aspoň niečo ak to aj bude NaN, to sa porieši v sql alebo pbi
#export do csv
df_melted.to_csv("households_expenditure_final.csv", index=False)
```
